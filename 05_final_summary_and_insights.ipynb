# Project Overview

This project analyzes the Brazilian Olist e-commerce dataset to uncover insights into sales trends, customer behavior, and operational performance using Python-based data analysis and visualization techniques.

# Key Findings

**Sales Insights**

- Sales show seasonal peaks (especially Q4).
- Revenue is concentrated in a few product categories.
- Sales are geographically concentrated in specific states.

**Customer Insights**

- Majority customers are one-time buyers.
- Top 20–30% customers contribute majority revenue.
- RFM analysis identified high-value “Champion” customers.

**Operational Insights**
  
- Late deliveries significantly reduce review scores.
- Delivery performance directly impacts customer satisfaction.
- Logistics improvement can increase ratings and retention.

# Actionable Recommendations

- Implement loyalty programs to convert one-time buyers into repeat customers.
- Focus marketing on high-value RFM segments.
- Improve logistics in high-delay regions.
- Optimize inventory for top-performing categories.
- Monitor delivery performance as a KPI tied to customer satisfaction.

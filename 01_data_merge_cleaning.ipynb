# Data Cleaning 

Converting date columns to datetime in orders

```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')
```

```python
orders['order_status'].value_counts()
```

The orders dataset does not contain any canceled or unavailable orders.

```python
orders.dtypes
```

order_items already has price (product price) and freight_value (shipping cost). Total item values needs to be calculated. A "total_item_price" column is added.

```python
order_items['total_item_price'] = (
    order_items['price'] + order_items['freight_value']
)
```

```python
order_items
```

total_item_price is created (product price + freight)

```python
products['product_category_name']
```

```python
products['product_category_name'].isna().sum()
```

The product_category_name column contains 610 missing values which needs to be handled

```python
products['product_category_name'] = (
    products['product_category_name'].fillna('unknown')
)
```

```python
products['product_category_name'].isna().sum()
```

The missing values replaced with 'unknown'

Product categories needs to be translated to English by merging with the
product_category_name_translation dataset.

```python
products = products.merge(
    product_category_name_translation,
    on='product_category_name',
    how='left'
)
```

```python
products.head()
```

```python
# Handling missing translations 
products['product_category_name_english'] = (
    products['product_category_name_english']
    .fillna('unknown')
)
```

```python
products['product_category_name_english'].isna().sum()
```

product merged with product_category_name and missing values replaced with "unknown"

```python
customers[['customer_state', 'customer_city']].isna().sum()
```

The customers dataset already contains customer_state and customer_city.
No missing values were found, so no cleaning required

# Merging the datasets

```python
df = order_items.merge(
    orders[['order_id', 'customer_id', 'order_purchase_timestamp']],
    on='order_id',
    how='inner'
)
```

```python
df = df.merge(
    customers[['customer_id', 'customer_state']],
    on='customer_id',
    how='left'
)
```

```python
df = df.merge(
    products[['product_id', 'product_category_name_english']],
    on='product_id',
    how='left'
)
```

```python
df.head()
```

Order items were merged with orders, customers, and products datasets.

Each row in the final dataset represents a single product item sold,
its order information, customer state, product category,
item price with freight value, and purchase timestamp.
